```python
# Load in our libraries
import pandas as pd
import numpy as np
import xgboost as xgb
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV

%matplotlib inline

# Import models
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier, ExtraTreesClassifier
```

# Feature Engineering

## Read the CSV

```python
data=pd.read_csv("mushroom_train.csv")
```

## Visualize the data

```python
pd.options.display.max_columns = 999
data.head()
```

## Map output signals to numerical quantities

```python
data["class"]=data["class"].map({"p":0,"e":1})
```

## Split the dataset into predictor and target variables

```python
train=data.iloc[:,1:]
target=data.iloc[:,0]
```

```python
train.head()
```

## Use One-Hot-Encoding to get dummy features

Since feature columns have categorical values we have to convert them to numerical entities in order to feed them
to an algorithm . The two options are either using Label Encoder or One Hot Encoder. As the amount of categorical
data is quite large , using Label Encoder will create nonsensical relationships among feature columns .
For Ex :- If odor_s is labelled as 0 and and order_p is labelled as 1 , then most of the algorithms will treat
odor_p > odor_s which doesn't make sense. Hence One Hot Encoding is to be preferred .

```python
train=pd.get_dummies(train)
```

```python
train.head()
```

# Selection of models

Since the number of feature columns after One Hot Encoding is quite large , hence bagging or bootstrapping models
are to be preferred over linear models such as Logistic Regression and Support Vector Classifier in order to avoid
the 'Curse of Dimensionality' .

```python
classifiers={"Random_Forest":RandomForestClassifier(),
             "XGB":xgb.XGBClassifier(),
             "AdaBoost":AdaBoostClassifier(),
             "Gradient_Boosting":GradientBoostingClassifier(),
             "Extra_Trees":ExtraTreesClassifier()
            }
```

## Validation of models

Split the data into a training and test set in order to evaluate the performance of the models.

```python
x_train,x_test,y_train,y_test=train_test_split(train,target,random_state=1,test_size=0.)
```

```python
for name,clf in classifiers.items():
    clf.fit(x_train,y_train)
    print name,clf.score(x_test,y_test)
```

## Final Validation

As the accuracies are quite high it doesn't hurt to re-evaluate your selection of model with a better evaluation
algorithm to avoid chances of overfitting.

```python
for name,clf in classifiers.items():
    print name,cross_val_score(clf, train, target, cv=5).mean()
```

## Final Model

Final model will be an ensemble of the Extreme Gradient Boosting , Extra Trees and Random Forests Classifiers for improving accuracy . Ensemble techniques often leads to a reduction in bias and variance .

# Final Model Creation

It is an ensemble of Random Forest , XGBoost and Extra Trees Classifiers . The predictions are based on the rounding off of the probabilities with which each individual classifier predicts whether an instance is class 1(edible) or not . The average probability of the 3 classifiers is  considered for prediction and since this is a binary classification task , rounding off the average probability will give the target output signal.

```python
# The final ensemble model
class model(object):
    
    def __init__(self):
        
        #Random Forest Classifier
        self.rf_clf= RandomForestClassifier(n_estimators=200,
                                           criterion="gini",
                                           max_depth=None,
                                           max_features="sqrt")
        
        #XGB Classifier
        self.xgb_clf = xgb.XGBClassifier(learning_rate =0.05,
                                         n_estimators=1000,
                                         max_depth=4,
                                         min_child_weight=6,
                                         gamma=0,
                                         subsample=0.8,
                                         colsample_bytree=0.8,
                                         reg_alpha=0.005,
                                         objective= 'binary:logistic',
                                         nthread=4,
                                         scale_pos_weight=1,
                                         seed=27)
        
        #Extra Trees Classifier
        self.et_clf = ExtraTreesClassifier(n_estimators=100,
                                           max_depth=None,
                                           max_features="sqrt")
    
    # Function to fit the classifiers on the dataset
    def fit(self,train,target):
        
        self.rf_clf.fit(train,target)
        self.xgb_clf.fit(train,target)
        self.et_clf.fit(train,target)
    
    # Function to predict the test dataset
    def predict(self,test):
        
        # Probabilities with which the classifiers predict class 1 type instance i.e edible
        rf_pred = self.rf_clf.predict_proba(test)[:,1]
        xgb_pred = self.xgb_clf.predict_proba(test)[:,1]
        et_pred = self.et_clf.predict_proba(test)[:,1]
        
        predictions = (rf_pred + xgb_pred + et_pred)/3.0
        predictions = np.round(predictions).astype(int)
        
        return predictions
    
    # Function to check the accuracy of the entire model
    def score(self,test,target):
        
        # Probabilities with which the classifiers predict class 1 type instance i.e edible
        rf_pred = self.rf_clf.predict_proba(test)[:,1]
        xgb_pred = self.xgb_clf.predict_proba(test)[:,1]
        et_pred = self.et_clf.predict_proba(test)[:,1]
        
        predictions = (rf_pred + xgb_pred + et_pred)/3.0
        predictions = np.round(predictions).astype(int)
        accuracy = np.sum(predictions==target).astype(float)/len(predictions)
        
        return accuracy
        
        
```

## Evaluating the final model

```python
fm = model()
fm.fit(x_train,y_train)
fm.score(x_test,y_test)
```
